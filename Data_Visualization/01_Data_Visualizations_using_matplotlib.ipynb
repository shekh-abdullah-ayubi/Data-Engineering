## __Data Visualizations with Matplotlib__
To gain more insights or understand the problem you're solving, it is very important to visualize the dataset that you are working with.

Matplotlib is powerful visualization tool in Python. It can be used to create 2D and 3D figures.

__Basics of Matplotlib__

```python
import matplotlib.pyplot as plt
import numpy as np

# If using notebook, we need this to display plots

%matplotlib inline 
```

A Simple Starter plot

```python
x_data = [1,2,3,4,5,6,7,8,9]
y_data = np.exp(x_data)
```

```python
plt.plot(x_data, y_data)
plt.title('This is a title')
plt.xlabel('X_data_label')
plt.ylabel('Y_data_label')
```

`x_data` and `y_data` can be lists, NumPy arrays, or a Pandas dataframes. In the latter parts, you will work with real world datasets where you can plot a feature as a dataframe.

Below is the anatomy of the figure. It is taken from and built with `matplotlib.org`.



__Subplots__\
`plt.subplot` will display multiple plots on same figure.

```python
# plt.subplot(number of rows, number of columns, plot number)

# Create emply figure with multiple plots
plt.subplot(1,3,1)
plt.plot(x_data, y_data)

plt.subplot(1,3,2)
plt.plot(x_data, np.sin(y_data))


plt.subplot(1,3,3)
plt.plot(x_data, np.cos(y_data))
```

__How `plt.subplot()` and subsequent calls work__

e.g. `plt.subplot(1, 3, 1)`:
* `1`: This indicates that the subplot grid will have 1 row.
* `3`: This indicates that the subplot grid will have 3 columns.
* `1`: This specifies that the current plot being created will be placed in the first position within this `1x3` grid. In a 1x3 grid, the positions are numbered from left to right, so '1' refers to the leftmost subplot.

__Figure Size__\
You can use `plt.figure(figsize=(width, height)` to adjust the figure size to your desied size.

```python
plt.figure(figsize=(5,3))

plt.plot(x_data, y_data)
```

You can also specify the dot per inch (DPI)

```python
plt.figure(figsize=(8,5), dpi=100)

plt.plot(x_data, y_data)
```

`figsize=(width, height)` can also be used in subplots that we saw earlier.

`plt.subplots(figsize=(10,5))`

__Saving Figure__\
We can save the Matplot figures in different formats, be it PNG, JPG, PDF, etc..

```python
figure = plt.figure(figsize=(8,5))
figure.savefig('expo-graph.png')
```

__Figure Title, Axis Labels and Legends__

```python
# A Title can be added to figure with plt.tiitle('.....')

x = np.linspace(0, 50, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.plot(x, y1)
plt.title('Harmonic motion')
```

```python
# X and Y labels can be added to  figure by plt.xlabel('...') or plt.ylabel('..')

plt.plot(x, y2)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
# Legend is added to figure by plt.legend()
plt.plot(x, y1, label = 'Y1')
plt.plot(x, y2, label = 'Y2')
plt.xlabel('Speed along X axis')
plt.ylabel('Sine and Cosine intersection')
plt.legend(loc='lower left')
```

__Text and Annotation__\
If you want to add texts or annotation in visualization, Matplotlib can allow you to do that.

```python
x = [1,2,3,4,5,6,7,8]
y = np.square(x)

# Plot x and y
plt.plot(x, y)

# Add the text to the plot
# plt.text(x position, y position, string, color or style)

plt.text(2,60, 'This texts is added at the plot where x=2, y=60', color='r')
```

```python
# Text Annotation

plt.plot(x, y)

plt.annotate('This is the Minimum point', 
             xy=(1,1), 
             xytext=(1.5, 30), arrowprops=dict(facecolor='red'))

plt.annotate('This is the Maximum point', # string to annotate
             xy=(8,64), # the position of x and y
             xytext=(1, 55), # the xy position to add the texts
             bbox=dict(boxstyle='round'), # or boxstyle can be 'round'
             arrowprops=dict(arrowstyle='fancy',connectionstyle='angle3, angleA=0, angleB=-90'))


plt.annotate('Middle, estimated', # string to annotate
             xy=(4.5,22), # the position of x and y
             xytext=(5, 12), # the xy position to add the texts
             bbox=dict(boxstyle='round', fc='none'), # or boxstyle can be 'round'
             arrowprops=dict(arrowstyle='wedge, tail_width=0.5', alpha=0.3))
```

There are alot of interesting things that you can do with annotation and texts by `plt.text()` and `plt.annotate()`. Learn more at [Matplotlib documentation](https://matplotlib.org/stable/tutorials/text/annotations.html#sphx-glr-tutorials-text-annotations-py).

How do we create customized plots?

__Object Oriented Method__\
If you want to create and customize figures, object oriented method is for you.

Below are examples

```python
# Create empty figure
fig = plt.figure()

# Create axes

ax = plt.axes()

# Plot on that axes

ax.plot(x, y, 'g') # green color

# Add the title 

ax.set_title('Square Graph')

# Add X and Y label

ax.set_xlabel('X axis Label')
ax.set_ylabel('Y axis Label')
```

This is the same as for `subplots`.

```python
# Create empty figure
#fig = plt.figure()

# Create axes
fig, axes = plt.subplots(nrows=1, ncols=2)

# Plot on that axes

axes[0].plot(x, y, 'g') # green color
axes[1].plot(x, y, 'b') # blue color

# Add the title 

axes[0].set_title('Subset 1')
axes[1].set_title('Subset 2')
```

You can also iterate through the axes of subplots. This would be very helpful if you have many rows and columns.

When using `axes` and subplots, the elements of figure can overlap. You can use `plt.tight_layout()` to correct that automatically.

```python
fig, axes = plt.subplots(nrows=1, ncols=3)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_title('Square graph')

# Show the figure

fig

# Use plt.tight_layout to adjust the positions of the axes, and labels automatically 
# Comment it and see how bad the figure is
plt.tight_layout()
```

To conclude the object oriented interface, you could use the function `ax.set()` and include all the plot things immediatey like `xlim`, `ylim`, `xlabel`, `ylabel`, `title`.

```python
# Create an axis 
# End the function with ; to get rid of unused figure details

ax = plt.axes()

ax.plot(x, y, 'g')
ax.set(title='Square graph with limits', 
       xlabel='X axis', 
       ylabel='Y axis',
       ylim=(-10,10), # This is x limit..Set by ax.set_xlim()
       xlim=(0,25)); # This is y limit..Set by ax.set_ylim() 
```
